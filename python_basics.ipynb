# Data Structures

 ##  Strings

Strings are a sequence of characters. They are used to store text like names, words, or sentences.

```python
name = "Python"

print(name)
print(len(name))
print(name.upper())
print(name.lower())
```

##  Lists

Lists store multiple items in a single variable. They are ordered and changeable (mutable).

```python
fruits = ["apple", "grapes", "mango"]

print(fruits)
print(fruits[0])
print(fruits[-1])

fruits.append("orange")
print(fruits)
```

##  Tuples

Tuples are ordered collections like lists, but they are immutable (cannot be changed).

```python
colors = ("red", "green", "blue")

print(colors)
print(colors[0])
print(colors[-1])
```

##  Sets

Sets are unordered collections of unique elements. They automatically remove duplicates.

```python
numbers = {1, 2, 3, 3, 2, 4}

print(numbers)
```

##  Dictionaries

Dictionaries store data in key-value pairs. They are used to map information.

```python
student = {
    "name": "Purva",
    "age": 18,
    "course": "Data Science"
}

print(student)
print(student["name"])
print(student["age"])

student["city"] = "Pune"
print(student)
```

# Day 4 - Loops, Conditions & Functions

Basic control flow, loops, functions, and mini practice tasks.

##  If Else Statements

If-else statements are used to make decisions in Python.

```python
number = 10

if number > 0:
    print("Positive number")
else:
    print("Negative number or zero")
```

##  Nested If Statements

Nested if statements are used when one condition exists inside another condition.

```python
age = 20

if age >= 18:
    if age >= 21:
        print("Eligible for driving and adult activities")
    else:
        print("Eligible for driving only")
else:
    print("Not eligible")
```

##  For Loop

A for loop is used to repeat a block of code multiple times.

```python
for i in range(5):
    print(i)
```

##  While Loop

A while loop keeps running as long as the condition is true.

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

##  Mini Task - Multiplication Table

This program prints the multiplication table of a number using a for loop.

```python
number = 5

for i in range(1, 11):
    print(number, "x", i, "=", number * i)
```

# Functions

##  Functions in Python

Functions are reusable blocks of code used to perform specific tasks.

```python
def greet():
    print("Hello, welcome to Python!")

greet()
```

##  Function Parameters

Parameters allow functions to accept input values.

```python
def greet(name):
    print("Hello", name)

greet("Purva")
greet("Python")
```

##  Return Statement

The return statement sends a value back from a function.

```python
def add(a, b):
    return a + b

result = add(5, 3)

print(result)
```

##  Lambda Functions

Lambda functions are short anonymous functions written in one line.

```python
square = lambda x: x * x

print(square(4))
print(square(7))
```

##  Mini Task - Grade Calculator Function

This function calculates grades based on marks.

```python
def grade_calculator(marks):

    if marks >= 90:
        return "A"

    elif marks >= 75:
        return "B"

    elif marks >= 50:
        return "C"

    else:
        return "Fail"


print(grade_calculator(95))
print(grade_calculator(70))
print(grade_calculator(40))
```
